```python
from catboost import CatBoostClassifier
```

```python
df_wine_test = pd.read_csv('Wine_TEST', header=None)
```

```python
names = {0:"type_of_wine"}
for i in list(df_wine_test.columns):
  if i != 0:
    names.update({i:f"proba_{i}"})
df_wine_test.rename(columns=names, inplace=True)
```

```python
model = CatBoostClassifier(iterations=100,
                           learning_rate=1,
                           depth=4)
eval_data = df_wine_test.drop(columns=["type_of_wine"])
# Fit model
model.fit(df2.drop(columns=["type_of_wine"]), df2.type_of_wine, None)
# Get predicted classes
preds_class = model.predict(eval_data)
# Get predicted probabilities for each class
preds_proba = model.predict_proba(eval_data)
# Get predicted RawFormulaVal
preds_raw = model.predict(eval_data, prediction_type='RawFormulaVal')
```

```python
confusion_matrix(preds_class, df_wine_test.type_of_wine)
```
